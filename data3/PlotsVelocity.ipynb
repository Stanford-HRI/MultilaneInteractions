```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style('white')
import ast
from __future__ import division
from collections import defaultdict
```

```python
filename = "velocity_results.txt"
data1, data2, data3, data4, data5, data6, data7, data8, data9 = [], [], [], [], [], [], [], [], []
capacity_data, capacity_average_data = defaultdict(list), {}
step_number, num_iter, robot_index = 0, 0, 0
with open(filename, 'r') as f:
    for line in f:
        res_str = line
        dict1 = ast.literal_eval(res_str)
        if num_iter % 20 == 0:
            step_number += 1
        velocity_temp = dict1['velocity']
        if dict1['isRobot']:
            if robot_index == 0:
                data1.append(velocity_temp)
                robot_index += 1
            elif robot_index == 1:
                data2.append(velocity_temp)
                robot_index += 1
            elif robot_index == 2:
                data3.append(velocity_temp)
                robot_index += 1
            elif robot_index == 3:
                data4.append(velocity_temp)
                robot_index += 1
            elif robot_index == 4:
                data5.append(velocity_temp)
                robot_index += 1
            elif robot_index == 5:
                data6.append(velocity_temp)
                robot_index += 1
            elif robot_index == 6:
                data7.append(velocity_temp)
                robot_index += 1
            elif robot_index == 7:
                data8.append(velocity_temp)
                robot_index += 1
            elif robot_index == 8:
                data9.append(velocity_temp)
                robot_index = 0
```

```python
plt.rc('font',**{'family':'serif','serif':['Palatino']})
plt.rc('text', usetex=True)
```

```python
plt.figure(figsize=(10,10))
# plt.plot(np.linspace(0,199,200), data1, label='data1', linewidth = 4.5)
plt.plot(np.linspace(0,125,126), data3[:126], '--',label='Autonomous Car 1', linewidth = 4.5, alpha = 0.7, color = 'purple')
# plt.plot(np.linspace(0,199,200), data4, label='data4', linewidth = 4.5)
plt.plot(np.linspace(0,125,126), data5[:126], label='Autonomous Car 2', linewidth = 4.5, alpha = 0.7, color = 'red')
# plt.plot(np.linspace(0,199,200), data8, label='data9', linewidth = 4.5)

plt.annotate("Local Interaction", xy=(65, 0.12), xytext=(80, 0.15), size=21, 
            arrowprops=dict(facecolor='black', shrink=0.1),
            )
plt.annotate('Local Interaction', xy=(36, 0.35), xytext=(10, 0.24), size=21, 
            arrowprops=dict(facecolor='black', shrink=0.1),
            )
sns.despine()
plt.xlabel("Number of Iterations of Nested Optimization", size=21)
plt.ylabel("Velocity", size=21)
plt.legend(prop={'size': 21})
plt.savefig("velocity_fig.pdf")
```
